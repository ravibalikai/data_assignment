## 6 marks

```python
marks = int(input('Enter the marks:'))
if marks > 90:
    print('Grade: A')
elif marks > 75 and marks < 89:
    print('Grade: B')
elif marks > 50 and marks < 74:
    print('Grade: C')
else:
    print('Fail')
```

##  7 positive or negative

```python
num = int(input('Enter the number:'))

if num > 0:
    if num % 2 == 0:
        print('positive and even')
    else:
        print('positive and odd')
elif num == 0:
    print('its zero')
else:
    print('its negative')
```
