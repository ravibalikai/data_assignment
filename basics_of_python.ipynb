```python
# Prime number

num=int(input('Enter the number'))
count = 0

for i in range(1,num+1):   #checking the num is divisible by i 
    if num % i == 0:
        count+=1
if count == 2:       # th eprime number sare divisble by only 1 and itself so if counts == 2 then  its prime
    print('Its a prime number')
else:
    print('not a prime number')
```

```python
print('Printing squares of Even numbers')
for i in range(100, 201):               # for loop for the range of 100 to 200. 
    if i % 2 == 0:
        square = i ** 2                  #  squares of the number
        print(f"{i}² = {square}")
```
